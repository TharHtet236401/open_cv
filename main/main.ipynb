```python
import pandas as pd
import numpy as np
import os
from glob import glob

import cv2
import matplotlib.pylab as plt

root_path = os.path.join(os.getcwd())
print(root_path)
cat_path = os.path.join(root_path, 'cats')
dog_path = os.path.join(root_path, 'dogs')

print(cat_path) 
print(dog_path)
```

```python
cat_images = glob(os.path.join(cat_path, '*.jpg'))
dog_images = glob(os.path.join(dog_path, '*.jpg'))

cat_photo = cat_images[20]
dog_photo = dog_images[20]

print(cat_photo)
print(dog_photo)

cat_photo_cv2 = cv2.imread(cat_photo)
cat_photo_mlt = plt.imread(cat_photo)

print(cat_photo_cv2.shape)
print(cat_photo_mlt.shape)



```

```python
#Display the image
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(cat_photo_cv2) # read with BGR
ax[0].set_title('Cat  with BGR')
ax[1].imshow(cat_photo_mlt) # read with RGB
ax[1].set_title('Cat  with RGB')
plt.show()

converted_cv2_rgb = cv2.cvtColor(cat_photo_cv2, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(converted_cv2_rgb)
ax[0].set_title('Cat  with RGB converted')
ax[1].imshow(cat_photo_mlt)
ax[1].set_title('Cat  with RGB')
plt.show()
```

```python
gray_cat_photo = cv2.cvtColor(cat_photo_cv2, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(gray_cat_photo, cmap='gray')
ax[0].set_title('Cat  with RGB converted')
ax[1].imshow(cat_photo_mlt)
ax[1].set_title('Cat  with RGB')
plt.show()
```

```python
# Resize the image
resized_cat_photo = cv2.resize(cat_photo_mlt, (200, 200))

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(resized_cat_photo)
ax[0].set_title('Cat  with RGB converted')
ax[1].imshow(cat_photo_mlt)
ax[1].set_title('Cat  with RGB')
plt.show()


resized_cat_photo.shape


```

```python
sharpen_kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]]);
blur_kernel = np.array([[0.04, 0.04, 0.04], [0.04, 0.04, 0.04], [0.04, 0.04, 0.04]])


sharpened_cat_photo = cv2.filter2D(cat_photo_mlt, -1, sharpen_kernel)
blurred_cat_photo = cv2.filter2D(cat_photo_mlt, -1, blur_kernel)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(sharpened_cat_photo)
ax[0].set_title('Cat  sharpened')
ax[1].imshow(blurred_cat_photo)
ax[1].set_title('Cat  blurred')
plt.show()

plt.imsave('cat_photo_sharpened.png', sharpened_cat_photo)
plt.imsave('cat_photo_blurred.png', blurred_cat_photo)
```
